### 丟一個銅板，丟了100次，出現正面 50 次的機率有多大。
(提示: 先想是哪一種分配，然後透過 python 語法進行計算)


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics

low=0
high=1
r = np.arange(low,high)

print(r)
X = stats.randint.rvs(low,high,size=50)
plt.hist(X,bins=13)
plt.show()
```
